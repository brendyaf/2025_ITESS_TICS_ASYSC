## CASO 1

```python
import numpy as np
import matplotlib.pyplot as plt

# ===== Caso 1 =====
R = 10           # ohm
L = 10e-3        # H
C = 25e-9        # F
Vm = 5
f = 1000
beta = 0         # grados

# Funcion de entrada
fs = 48000
T = 0.02
N = int(fs * T)
t = np.linspace(0, T, N, endpoint=False)
```

```python
# señales
omega0 = 2*np.pi*f
vin = Vm * np.cos(omega0 * t + np.deg2rad(beta))

# H(jw)
freqs = np.fft.rfftfreq(N, 1/fs)
omega = 2*np.pi*freqs
H = 1.0 / (1.0 - (omega**2)*L*C + 1j*omega*C*R)

# FFT entrada->salida en frecuencia y reconstrucción en tiempo
Vin = np.fft.rfft(vin)
Vout = H * Vin
vout = np.fft.irfft(Vout, n=N)
```

```python
# --- Plots ---
plt.figure(figsize=(10,4))
plt.plot(t, vin, label='vin(t)', alpha=0.6)
plt.plot(t, vout, label='vout(t)', linewidth=0.9)
plt.title("Caso 1 - Respuesta temporal (vin y vout)")
plt.xlim(0, 5/f)
plt.xlabel("Tiempo (s)"); plt.ylabel("Voltaje (V)")
plt.legend(); plt.grid(); plt.show()

plt.figure(figsize=(10,3))
plt.plot(freqs, np.abs(Vout))
plt.title("Caso 1 - Espectro (magnitud) de la salida")
plt.xlim(0, fs/2); plt.xlabel("Frecuencia (Hz)"); plt.ylabel("|Vout(f)|")
plt.grid(); plt.show()

plt.figure(figsize=(10,3))
plt.plot(freqs, np.angle(Vout, deg=True))
plt.title("Caso 1 - Fase del espectro de la salida")
plt.xlim(0, fs/2); plt.xlabel("Frecuencia (Hz)"); plt.ylabel("Fase (°)")
plt.grid(); plt.show()
```
