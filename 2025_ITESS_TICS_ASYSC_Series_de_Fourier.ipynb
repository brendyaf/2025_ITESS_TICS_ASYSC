## ðŸŒ¸ Serie de Fourier

La serie de Fourier permite representar una funciÃ³n periÃ³dica \( f(t) \) como una suma infinita de senos y cosenos.

---

### ðŸ“˜ ExpresiÃ³n general

$$
S(t) = a_0
$$

$$
s(f) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(2\pi f_n t) + b_n \sin(2\pi f_n t)Â \right]
$$

---

### ðŸ§® Coeficientes de la serie

Los coeficientes estÃ¡n definidos como:

$$
a_0 = \frac{1}{T} \int_{0}^{T} f(t)\, dt
$$

$$
a_n = \frac{2}{T} \int_{0}^{T} f(t)\cos(2\pi f_n t)\, dt
$$

$$
b_n = \frac{2}{T} \int_{0}^{T} f(t)\sin(2\pi f_n t)\, dt
$$



âœ¨ Donde \( T \) es el periodo fundamental y \( f_n \) representa las frecuenciasÂ armÃ³nicas.


```python
from thinkdsp import CosSignal, SinSignal
pi=3.1416
k=1
f=0.25

## a1
st = CosSignal(freq=f, amp=(2*k)/pi, offset=0)

## a3
st2 = st + CosSignal(freq=f*3, amp=-(3*k)/(3*pi), offset=0)

## a5
st3 = st2 + CosSignal(freq=f*5, amp=(2*k)/(5*pi), offset=0)
```

## Wave: Audio 1
A Signal represents a mathematical function defined for all values of time. If you evaluate a signal at a sequence of equally-spaced times, the result is a Wave. framerate is the number of samples per second.

```python
from thinkdsp import read_wave
import matplotlib.pyplot as plt

wave = read_wave(r'C:\Users\Brandy\Music\Audios\Hola_Como-estas.wav')

wave.plot()
decorate(xlabel='Time (s)')
plt.show()
```

```python
##IPython provides an Audio widget that can play a wave.

from IPython.display import Audio
audio = Audio(data=wave.ys, rate=wave.framerate)
audio
```

```python
## Wave also provides `make_audio()`, which does the same thing:
wave.make_audio()
```

## Spectrums
Wave provides make_spectrum, which computes the spectrum of the wave.



```python
# Make a spectrum object
## Audio 1
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(ylabel='Audio 1')
decorate(xlabel='Frequency (Hz)')
plt.show()

## Audio 2
spectrum2 = wave2.make_spectrum()
spectrum2.plot()
decorate(ylabel='Audio 2')
decorate(xlabel='Frequency (Hz)')
plt.show()
```

```python
##Audio 1
spectrum.plot(high=10000)
decorate(xlabel='Frequency (Hz)')
## Audio 2
spectrum2.plot(high=10000)
decorate(xlabel='Frequency (Hz)')
```

We can convert the filtered spectrum back to a wave:

```python
## Audio 1
filtered = spectrum.make_wave()
## Audio 2
filtered2 = spectrum2.make_wave()
```

And then normalize it to the range -1 to 1.

```python
## Audio 1
filtered.normalize()
## Audio 2
filtered2.normalize()
```

Before playing it back, I'll apodize it (to avoid clicks).

```python
## Audio 1
filtered.apodize()
filtered.plot()
## Audio 2
filtered2.apodize()
filtered2.plot()
decorate(xlabel='Time (s)')
```

And I'll do the same with the original segment.

```python
## Audio 1
segment.normalize()
segment.apodize()
segment.plot()
## Audio 2
segment2.normalize()
segment2.apodize()
segment2.plot()
decorate(xlabel='Time (s)')
```
